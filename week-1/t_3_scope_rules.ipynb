# Scopes in Python

Scope in python determines where in your code the variables are:
- Visible and
- Accessible

There are 4 levels of scope in python:
1. Local
2. Enclosing
3. Global
4. Built-in


### What is `namespace`?

Namespace is the dictionary that maps the object and determines their scope.

We can use builtin methods like `globals()` and `locals()` to check the global and local variables

```python
x = 10 # global variable

def func():
    y = 20
    x = 300 # local variable

    print(f"globals() response: {globals()}")
    print(f"locals() response: {locals()}")

    global_vars = globals()
    print(f"type of global_vars: {type(global_vars)}")
    print(f"global_vars['x']: {global_vars['x']}")

    local_vars = locals()
    print(f"type of local_vars: {type(local_vars)}")
    print(f"local_vars['y']: {local_vars['y']}")
    print(f"local_vars['x']: {local_vars['x']}")

func()
```

We can clearly see above that `x` inside the function is local variable and `x` outside the function is global variable
